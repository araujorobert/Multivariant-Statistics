```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('companies79.csv')
data = data.drop(['Unnamed: 0'], axis=1)
data.head()
```

```python
# Guardamos todas las variables numericas en la variable X
X = data[['V2','V3','V4','V5','V6','V7']]
```

```python
# Haciendo un scatter plot matrix
sns.pairplot(X)
plt.show()
```

```python
columnas = ['V2','V3','V4','V5','V6','V7','V8']
X2 = data[columnas]

X2.boxplot(by='V8', vert=0, figsize=(16,10))
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.distplot(X2.iloc[:,0])
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.distplot(X2.iloc[:,1])
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.distplot(X2.iloc[:,2])
plt.show()
```

```python
plt.figure(figsize=(20,6))
sns.distplot(X2.iloc[:,1])
sns.distplot(X2.iloc[:,2])

sns.distplot(X2.iloc[:,3])

sns.distplot(X2.iloc[:,4])

sns.distplot(X2.iloc[:,5])


plt.show()
```

```python
# Haciendo un scatter plot matrix
sns.pairplot(X2, hue='V8')
plt.show()
```
